```python
import os
import pandas as pd
import numpy as np
Data = pd.read_csv('Toyota.csv',index_col=0,na_values=["??","????"])
```

```python
Data.head(50)
Data.fillna(0.0,inplace=True)
Data.head(50)

```

```python
Data.info()
Data.describe()
```

```python
Data['Automatic'] = Data['Automatic'].replace({0:'Non Automatic',1:'Automatic'})
Data.head(5)
```

```python
Data.rename(columns={'HP':'Horse Power'},inplace=True)
Data.head(5)
```

```python
Data['Horse Power'] = Data['Horse Power'].astype('int')
Data.tail(5)
```

```python
Data['FuelType'] = Data['FuelType'].astype('str')
Data['FuelType'] =Data['FuelType'].replace({'0.0':'No Fuel Type'})
Data.head(5)
Data.tail(5)
```

```python
Data.rename(columns={'MetColor':'Metallic Color'},inplace=True)
Data['Metallic Color'] = Data['Metallic Color'].astype('str')
Data['Metallic Color'] = Data['Metallic Color'].replace({'0.0':'No Metallic Color','1.0':'Metallic Color'})
Data['Age'] = Data['Age'].astype('int')
Data['KM'] = Data['KM'].astype('str')+ ' Km'
# Data['KM'] = Data['KM'].astype('int')
Data['Weight'] = Data['Weight'].astype('int')
Data['Doors'] = Data['Doors'].replace({'three':3,'four':4,'five':5}) 
Data.head(50)
Data.tail(50)
```

```python
Data.to_excel('Toyota1.xlsx',index=False)
```

```python
j = ['Automatic','FuelType','Metallic Color','Doors','Weight','KM','Age','Horse Power','Price','CC']
for i in j:
    # it will give the frequency of each category
    a = pd.crosstab(index=Data[i],columns='count',dropna=True)
    print(a)
```
